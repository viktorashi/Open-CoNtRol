```python
import tellurium as te

rr = te.loada('''
        S0 -> KS1; k1*S0
        KS1 -> S2; k2*KS1
        S2 + F -> FS2; k3*S2*F
        FS2 -> F; k4*FS2
        F -> S0 + F; k5*F

        F = 0.874108
        FS2 = 7.620157734
        KS1 = 7.620157734
        S0 = 7.270157734
        S2 = 0.6000000000

        k1 = 0.1329759342
        k2 = 0.1329759342
        k3 = 2
        k4 = 0.1329759342
        k5 = 1
''')
checked_species = ['S0', 'KS1']
checked_with_time = checked_species
checked_with_time.insert(0,'time')
rr.simulate(0, 1000, 1000, selections=checked_with_time)
rr.plot(title='My first plot', xlabel='Time', ylabel='specii', ordinates = checked_species)
```
