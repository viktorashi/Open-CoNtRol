Plotuim o variabila de celalata

```python
import tellurium as te

rr = te.loada('''
    A -> B; k1*A
    B -> C; k2*B
    A = 10; B = 0; C = 20;
    k1 = 0.1; k2 = 0.2; 
''')

rr.simulate(0, 10, 100, ['A', 'B'])
rr.plot(xlabel='A', ylabel = 'B')

rr.reset()
```

Plotuim 2 variabile in functie de timp fiecare

```python
import matplotlib.pyplot as plt

ax = plt.figure().add_subplot(projection='3d')
m = rr.simulate(0, 10, 100, ['time', 'A', 'B'])
ax.plot(m[:,1], m[:,2], m[:,0], label ='lmao sunt jemeker')
ax.legend()
ax.set_xlabel('A')
ax.set_ylabel('B')
ax.set_zlabel('time')
ax.view_init(elev=20., azim=-35, roll=0)

plt.show()
```
